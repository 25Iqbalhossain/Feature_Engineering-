# Just a example of own  function ordinal_encoding

```python
def ordinal_encode(data, category_order):
    """
    Encodes categorical data using ordinal encoding.

    Parameters:
    data (list): The input data to encode.
    category_order (list): Specifies the order of categories.

    Returns:
    encoded_data (list): Ordinally encoded data.
    """
    encoded_data = []
    for value in data:
        encoded_data.append(category_order.index(value))
    return encoded_data

# Example usage
data = ["low", "medium", "high", "medium"]
category_order = ["low", "medium", "high"]  # Specify category order

encoded_data = ordinal_encode(data, category_order)
print("Ordinally Encoded Data:")
print(encoded_data)
```
