# This line use for remove unnecessary column

```python
df=df.iloc[:,2:]
df.head()
```

# Always remember  before train the data please do train_test_split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
X_train
```

# Tranformer of OrdinalEncoding 

```python
from sklearn.preprocessing import OrdinalEncoder
```

if we didn't use manually categories its generate value randomly 

```python
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
```

# On this line of  code transform categorical value to numerical

```python
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
```
