# Now Applying Column Transformation Just do this thing In SingleLine

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1', StandardScaler(), ['age']),
    ('tnf2', SimpleImputer(), ['fever']),
    ('tnf3', OrdinalEncoder(), ['cough']),
    ('tnf4', OneHotEncoder(sparse_output=False, drop='first'), ['gender', 'city'])
], remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```

# Another Approach # 

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ('tnf1', StandardScaler(), ['age']),
    ('tnf2', SimpleImputer(strategy='mean'), ['fever']),
    ('tnf3', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OrdinalEncoder())
    ]), ['cough']),
    ('tnf4', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(sparse_output=False, drop='first'))
    ]), ['gender', 'city'])
], remainder='passthrough')
```

```python
transformer.fit_transform(X_train)
```
